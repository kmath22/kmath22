```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head(5)
```

2. Now youâ€™re ready to set H4TO3 with values of 6 and 8 to blank. Copy and paste the code that does that.  Then copy and paste the new frequency table.  (Just run the entire code file.)

```python
df['Smoke_Reg'] = df['H4TO3']
df['Smoke_Reg'].value_counts()
```

```python
import researchpy as rp
rp.summary_cat(df['Smoke_Reg'])
```

```python
df['Smoke_Reg'] = df['Smoke_Reg'].loc[ df['Smoke_Reg'] != ' ']
df['Smoke_Reg'] = df['Smoke_Reg'].loc[ df['Smoke_Reg'] != '6']
df['Smoke_Reg'] = df['Smoke_Reg'].loc[ df['Smoke_Reg'] != '8']
df['Smoke_Reg'].value_counts()
```

3. Notice that your frequency table still has 7 -- the legitimate skips.  Recall that the people that got to skip this question said that they didnâ€™t smoke.  So we want change this to 0.  Copy and paste the code that does that.  Then copy and paste the new frequency table.


```python
df['Smoke_Reg'].replace( ['0','1','7'] , ['No', 'Yes', 'No'] , inplace = True) 
df['Smoke_Reg'].value_counts()
```

```python
rp.summary_cat(df['Smoke_Reg'])
```
